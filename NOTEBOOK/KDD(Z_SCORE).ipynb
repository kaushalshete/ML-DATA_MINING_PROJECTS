```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
iris_data = pd.DataFrame(data=iris.data, columns=iris.feature_names)
```

```python
iris_data.head()
```

```python
iris_data.describe()
```

```python
z_score = (iris_data - iris_data.mean())/ iris_data.std()
print(z_score)
```

```python
outliers = (np.abs(z_score) > 3)
print("Outlier points (rows where Z-score > 3 or < -3):")
iris_data[outliers.any(axis=1)]
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=z_score, palette='Set2')
plt.title("Box Plot of Z-scores for Iris Dataset Features")
plt.show()
```
