```python
import numpy as np
import pandas as pd
from google.colab import drive

# Mount Google Drive
drive.mount("/content/drive")

# File path
filepath = '/content/drive/My Drive/Colab Notebooks/KDD/Titanic-Dataset.csv'

# Load dataset
df = pd.read_csv(filepath)
df.head()

# Drop rows where 'Survived' is missing
df.dropna(subset=["Survived"], inplace=True)

# Fill missing values
df['Age'].fillna(df['Age'].mean(), inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)

# Encode categorical variables
df['Sex'] = df['Sex'].map({'male': 1, 'female': 0})
df['Embarked'] = df['Embarked'].map({'C': 0, 'S': 1, 'Q': 2})

# Feature engineering
df['FamilySize'] = df['SibSp'] + df['Parch']
df['IsAlone'] = df['FamilySize'].apply(lambda x: 1 if x == 0 else 0)

# Drop irrelevant columns
df.drop(['Name', 'Ticket', 'Cabin'], axis=1, inplace=True)

# Save cleaned data to new CSV
df.to_csv("titanic_transformed.csv", index=False)

# Reload and display to verify
d = pd.read_csv("titanic_transformed.csv")
display(d.head())
```
