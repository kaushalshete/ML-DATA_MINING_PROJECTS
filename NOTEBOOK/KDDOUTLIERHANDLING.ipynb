```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("/content/train_and_test2.csv")

# Initial data check
print(df.head())
print(df.isnull().sum())

# Fill missing values
df["Age"].fillna(df["Age"].mean(), inplace=True)
df["Cabin"].fillna('Unknown', inplace=True)
df["Fare"].fillna(df["Fare"].mean(), inplace=True)
df["Embarked"].fillna('S', inplace=True)

# Verify no missing values remain
print(df.isnull().sum())

# Plot Fare outliers
plt.figure(figsize=(10, 5))
plt.xlabel("Outliers")
plt.ylabel("Fare")
plt.title("Fare Outliers")
sns.boxplot(x=df["Fare"])
plt.show()

# Outlier removal in Age column using Z-score method
mean = df["Age"].mean()
std = df["Age"].std()
threshold = 3
outliers = (df["Age"] - mean).abs() > threshold * std
df_no_outliers = df[~outliers]

# Plot Age outliers before and after removal
plt.figure(figsize=[4,3])
plt.xlabel("Outliers")
plt.ylabel("Age")
plt.title("Age Outliers")
sns.boxplot(x=df["Age"])
plt.show()

plt.figure(figsize=[4,3])
plt.xlabel("Outliers")
plt.ylabel("Age (Cleaned)")
plt.title("Age Outliers Removed")
sns.boxplot(x=df_no_outliers['Age'])
plt.show()

# Optional: Re-check Fare outliers after filtering
plt.figure(figsize=[4,3])
plt.xlabel("Outliers")
plt.ylabel("Fare")
plt.title("Fare Outliers After Removing Age Outliers")
sns.boxplot(x=df_no_outliers["Fare"])
plt.show()
```
