```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import fpgrowth, association_rules
import pandas as pd

# Dataset
data = {
    'Transaction': ['T1', 'T2', 'T3', 'T4', 'T5', 'T6', 'T7', 'T8', 'T9'],
    'Items': [
        ['A', 'B', 'E'],
        ['B', 'D'],
        ['B', 'C'],
        ['A', 'B', 'D'],
        ['A', 'C'],
        ['B', 'C'],
        ['A', 'C'],
        ['A', 'B', 'C', 'E'],
        ['A', 'B', 'C']
    ]
}

# Convert data to transaction format
transactions = data['Items']

# Initialize TransactionEncoder
encoder = TransactionEncoder()

# Transform transaction data into one-hot encoded DataFrame
onehot = encoder.fit(transactions).transform(transactions)

# Convert one-hot encoded data to DataFrame
df = pd.DataFrame(onehot, columns=encoder.columns_)

# Finding frequent itemsets using FP-Growth
frequent_itemsets = fpgrowth(df, min_support=0.33, use_colnames=True)

# Generate association rules
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.5)

# Display results
print("Frequent Itemsets:")
print(frequent_itemsets.to_string(index=False))

print("\nAssociation Rules:")
print(rules.to_string(index=False))
```
